```python
import json
import os
import pandas as pd
pd.set_option('display.max_columns', 500)
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
records = pd.read_csv("external_output/afa_chatbot_records-CSIT-20200803_20211101.csv")
records["斷詞結果"] = records["斷詞結果"].apply(lambda x: str(x).replace("?","").replace(","," "))
records["關鍵短語抽取結果"] = records["關鍵短語抽取結果"].apply(lambda x: str(x).replace("?","").replace(","," "))
records["關鍵短語抽取結果"] = np.where(records["關鍵短語抽取結果"] == "nan", records["斷詞結果"], records["關鍵短語抽取結果"])
records[["month","weekOfYear","file_date","問法","關鍵短語抽取結果"]].to_csv("external_output/0211_update/afa_chatbot_records-CSIT-20200803_20211101.csv", index=False)
```

```python
sheets = pd.read_excel("external_output/original/afa_CSIT-phrase-stats-original-20200803_20211101.xlsx", 
                            sheet_name=["每月關鍵詞合併統計","每週關鍵詞合併統計","每日關鍵詞合併統計"])

bymonth = sheets["每月關鍵詞合併統計"]
byweek = sheets["每週關鍵詞合併統計"]
byday = sheets["每日關鍵詞合併統計"]

bymonth = bymonth[bymonth["count"] > 0].reset_index(drop=True)
byweek = byweek[byweek["count"] > 0].reset_index(drop=True)
byday = byday[byday["count"] > 0].reset_index(drop=True)
```

```python
with pd.ExcelWriter('external_output/0211_update/afa_CSIT-phrase-stats-original-20200803_20211101.xlsx') as writer:
    bymonth.to_excel(writer, sheet_name=f'每月關鍵詞合併統計')
    byweek.to_excel(writer, sheet_name=f'每週關鍵詞合併統計')
    byday.to_excel(writer, sheet_name=f'每日關鍵詞合併統計')
```
